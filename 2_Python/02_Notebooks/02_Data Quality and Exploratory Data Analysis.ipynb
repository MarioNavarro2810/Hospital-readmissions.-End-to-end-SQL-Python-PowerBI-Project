### 1.4. Data Inconsistencies

We detect logical inconsistencies within the dataset.  
For instance, a discharge date (`fecha_alta`) should never precede the admission date (`fecha_admision`).

```python
df.loc[df.fecha_alta < df.fecha_admision]
```

The length of stay in the intensive care unit should not exceed the total hospital stay, since the latter includes the former.

```python
df.loc[df.duracion_estancia_ci > df.duracion_estancia, 'id_admision':'duracion_estancia_ci']
```

```python
# Correct the values of duracion_estancia_ci when they exceed duracion_estancia
df.loc[df['duracion_estancia_ci'] > df['duracion_estancia'], 'duracion_estancia_ci'] = df['duracion_estancia']
```

```python
# Verify that the correction has been applied
df.loc[df.duracion_estancia_ci > df.duracion_estancia, 'id_admision':'duracion_estancia_ci']
```

## 2.2. EDA: Cathegorical values

We analyze the frequency distribution of categorical variables to understand the composition of the dataset.  
This step provides insights into variable balance and possible data dominance issues that could affect modeling.

```python
import matplotlib.pyplot as plt

for variable in df.select_dtypes('object').columns:
    frecuencias = df[variable].value_counts()
    frecuencias.plot.bar()
    plt.title(f'Frecuencia de valores en {variable}')
    plt.show();
```
