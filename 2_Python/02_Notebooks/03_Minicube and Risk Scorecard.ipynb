#### Prioritization of Significant Categorical Variables

To identify the most relevant categorical variables, we apply a **Chi-square test of independence** between each variable and the target (`genera_reingreso`).  
This test quantifies the statistical relationship between each feature and the likelihood of a future readmission.

A ranking is then generated based on the Chi-square statistic, allowing us to **prioritize variables** that show the strongest association with hospital readmissions.

```python
from scipy.stats import chi2_contingency
import pandas as pd

def chi_square_test(df, target_col, categorical_col):
    # Temporarily remove rows with NaN values in the current variable
    df_filtered = df.dropna(subset=[categorical_col])

    # Build the contingency table
    contingency_table = pd.crosstab(df_filtered[categorical_col], df_filtered[target_col])

    # Apply the Chi-square test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    return chi2, p, dof, expected

# Run the Chi-square test for each categorical variable
results = {}

for var in categoricas:
    chi2, p, dof, expected = chi_square_test(risk2, 'genera_reingreso', var)
    results[var] = {'chi2': chi2, 'p': p}

# Create a ranking based on the Chi-square value
ranking = pd.DataFrame.from_dict(results, orient='index').reset_index()
ranking.columns = ['Variable', 'Chi2', 'P-valor']
ranking = ranking.sort_values(by='Chi2', ascending=False).reset_index(drop=True)

# Print the ranking
print("Ranking de significaciÃ³n de variables categÃ³ricas:")
print(ranking)

```

```python
ranking[ranking['P-valor'] <= 0.01]
```

```python
# Since there are too many significant variables and a clear cutoff point, 
# we will manually select the top ones
significativas_cat = ranking.Variable[1:10].to_list()
significativas_cat
```

#### Prioritization of Significant Continuous Variables

We now apply **T-tests to continuous** variables in order to identify those that show statistically significant differences between patients who were readmitted and those who were not.

```python
df_continuas = risk2.loc[:, ~risk2.columns.isin(categoricas)]
df_continuas
```

```python
from scipy.stats import ttest_ind

# List to store significant continuous variables
significativas_con = []

# Iterate over each continuous variable
for col in df_continuas.columns:
    # Temporarily remove rows with NaN values in the current variable
    df_filtered = df_continuas.dropna(subset=[col])

    # Split into two groups based on 'genera_reingreso'
    group1 = df_filtered[risk2.loc[df_filtered.index, 'genera_reingreso'] == 0][col]
    group2 = df_filtered[risk2.loc[df_filtered.index, 'genera_reingreso'] == 1][col]
    
    # Check that both groups have enough data
    if len(group1) > 30 and len(group2) > 30:
        # Perform T-test
        t_stat, p_value = ttest_ind(group1, group2, equal_var=True)
        
        # Diagnosis

        if p_value < 0.01:  # Nivel de significancia del 1%
            diagnosis = "Significativa"
            significativas_con.append(col)  # Agregar a la lista de significativas
        else:
            diagnosis = "No significativa"
    else:
        t_stat, p_value, diagnosis = None, None, "No calculable (pocos datos)"
    
    print(f"Variable: {col}, T-statistic: {t_stat}, p-value: {p_value}, DiagnÃ³stico: {diagnosis}")

# Display the list of significant variables
print("\nVariables significativas:")
print(significativas_con)

```

Finally, we **combine our chosen significant categorical and continous variables**

```python
# Combine significant categorical and continuous variables

significativas = significativas_cat + significativas_con[:3]
significativas
```

##### Customization of the risk Score-card

Here we provide the **instructions for the hospital staff** to assign points according to their criteria.
The following function can be **customized by modifying the ifâ€“then rules to define how each patient characteristic contributes to the overall risk score.**

We create a **function called calculate_scoring**, which receives a single record from the dataset and returns a total risk score.

The score starts at zero, and based on the patientâ€™s values in each variable, points are added according to predefined ifâ€“then rules.

**These rules can be manually adjusted by hospital staff** to tailor the scoring criteria.

```python

def calcular_scoring(registro):
    scoring = 0  # Inicializamos el scoring en 0
    
# Rules to calculate the score based on the given table
    if registro.get("cardiomiopatia_previa") == 1:
        scoring += 52
    else:
        scoring += 8

    if registro.get("dm") == 1:
        scoring += 49
    else:
        scoring += 11

    if "duracion_estancia_ci_disc" in registro:
        duracion_estancia_ci_disc = registro["duracion_estancia_ci_disc"]
        if duracion_estancia_ci_disc in ["(-1, 0]", "(0, 10]"]:
            scoring += 19 if duracion_estancia_ci_disc == "(-1, 0]" else 23
        elif duracion_estancia_ci_disc == "(10, 20]":
            scoring += 30
        elif duracion_estancia_ci_disc == "(20, 28]":
            scoring += 42
        elif duracion_estancia_ci_disc == "(28, 35]":
            scoring += 27
        elif duracion_estancia_ci_disc == "(35, 100]":
            scoring += 33

    if "duracion_estancia_disc" in registro:
        duracion_estancia_disc = registro["duracion_estancia_disc"]
        if duracion_estancia_disc == "(0, 10]":
            scoring += 22
        elif duracion_estancia_disc == "(10, 25]":
            scoring += 29
        elif duracion_estancia_disc == "(25, 60]":
            scoring += 30
        elif duracion_estancia_disc == "(60, 100]":
            scoring += 100

    if "edad_disc" in registro:
        edad_disc = registro["edad_disc"]
        if edad_disc == "(0, 20]":
            scoring += 13
        elif edad_disc == "(20, 50]":
            scoring += 17
        elif edad_disc == "(50, 80]":
            scoring += 23
        elif edad_disc == "(80, 150]":
            scoring += 31

    if registro.get("erc") == 1:
        scoring += 51
    else:
        scoring += 10
    
    if registro.get("iamcest") == 1:
        scoring += 47
    else:
        scoring += 12
    
    if registro.get("icfen") == 1:
        scoring += 53
    else:
        scoring += 8
    
    if registro.get("icfer") == 1:
        scoring += 54
    else:
        scoring += 7
    
    if registro.get("insuficiencia_cardiaca") == 1:
        scoring += 54
    else:
        scoring += 6
    
    if registro.get("lra") == 1:
        scoring += 50
    else:
        scoring += 10
    
    if registro.get("sca") == 1:
        scoring += 47
    else:
        scoring += 12
    
    return scoring
```

```python
# Apply the function to each record in the DataFrame 'risk'
risk['scoring'] = risk.apply(calcular_scoring, axis=1)
```

#### 2.2.5. SCALE ADJUSTMENT

For the end user, it is preferable to present a scoring system that is easy to interpret.
Therefore, we will rescale our risk score to a range from 0 to 10,
where 0 represents the lowest possible readmission risk and 10 represents the highest.

This transformation ensures that the results are standardized and comparable across patients,
making the score more intuitive for clinicians and decision-making systems.

```python
# First, understand the current range

risk['scoring'].agg(['min', 'max'])
```

```python
# Calculate the minimum and maximum values achievable with our scoring table
min_val = risk['scoring'].min()
max_val = risk['scoring'].max()

# Transform the score to a 0â€“10 scale
risk['scoring'] = (risk['scoring'] - min_val) / (max_val - min_val) * 10
risk['scoring'] = risk['scoring'].astype('int')

risk['scoring'].agg(['min', 'max'])
```

```python
## Verify the new range
risk.scoring.value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

# Convert scoring to numeric (if not already)
risk['scoring'] = pd.to_numeric(risk['scoring'])

# Group by `scoring` and calculate the average readmission rate
scoring_avg = risk.groupby('scoring')['genera_reingreso'].mean().sort_index(ascending=True)

# Create the bar chart
plt.figure(figsize=(12, 6))
scoring_avg.plot(kind='bar', color='skyblue')

# Configure title and labels
plt.title('Media de genera_reingreso por puntuaciÃ³n de scoring', fontsize=16)
plt.xlabel('Scoring', fontsize=14)
plt.ylabel('Media de genera_reingreso', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```
